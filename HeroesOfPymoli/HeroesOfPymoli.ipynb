```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

# purchase_data.head()
```

```python
#calculate and display total number of unique players

total_players = purchase_data["SN"].nunique()
total_players

total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchasing_analysis = pd.DataFrame({
    "Number of Unique Purchases": [purchase_data["Item ID"].nunique()],
    "Average Price": [round(purchase_data["Price"].mean(),2)],
    "Number of Purchases": [len(purchase_data.index)],
    "Total Revenue": [purchase_data["Price"].sum()]
})

pd.options.display.float_format = '${:,.2f}'.format

purchasing_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
unique_purchasers = purchase_data.drop_duplicates(subset="SN")

gender_demographics = pd.DataFrame({
    "Gender" : unique_purchasers["Gender"].value_counts()
})

gender_demographics["Percentage"] = gender_demographics["Gender"]/total_players*100
gender_demographics['Percentage'] = gender_demographics['Percentage'].map('{:,.2f}%'.format)

gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# purchase_count = purchase_data["Item ID"].count()
# avg_price = purchase_data["Price"].sum()/purchase_count

grouped_purchase_data = purchase_data.groupby(['Gender'])

final_groups_df = pd.DataFrame([])
final_groups_df["Purchases"] = grouped_purchase_data["Purchase ID"].count()
final_groups_df["Purchase Value"] = grouped_purchase_data["Price"].sum()
final_groups_df["Unique Purchases"] = grouped_purchase_data["SN"].nunique()
final_groups_df["Avg Purchase PP"] = final_groups_df["Purchase Value"] / final_groups_df["Unique Purchases"]
final_groups_df["Average P Price"] = final_groups_df["Purchase Value"] / final_groups_df["Purchases"]

final_groups_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#bin age groups into new column in purchase data
bin_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
bin_var = [0,9,14,19,24,29,34,39,1000]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bin_var,  labels = bin_labels)

# purchase_data.head()
```

```python
#updating old unique_purchasers to include bins
unique_purchasers = purchase_data.drop_duplicates(subset="SN")

#group new dataframe according to age group
purchase_bins = unique_purchasers.groupby(['Age Group'])

#new dataframe to receive group calculations and display
purchase_bins_unique = pd.DataFrame([])

purchase_bins_unique["Total Count"] = purchase_bins["Age Group"].count()
purchase_bins_unique["Percentage of Players"]=(purchase_bins_unique["Total Count"]/total_players*100).map("{:.2f}%".format)

#display
purchase_bins_unique
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#group main dataframe according to age group
grouped_by_group = purchase_data.groupby(["Age Group"])

tot_value = grouped_by_group["Price"].sum()

tots_by_age = grouped_by_group.count()
tots_by_age = tots_by_age.drop(columns=["SN","Age","Gender","Item ID","Price","Item Name"]).rename(columns = {"Purchase ID":"Purchase Count"})
tots_by_age["Average Purchase Price"] = tot_value / tots_by_age["Purchase Count"]
tots_by_age["Total Purchase Value"] = tot_value
tots_by_age["Average Total Purchase Per Person"] = tot_value / purchase_bins_unique["Total Count"]

tots_by_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purch_count = purchase_data['Item ID'].groupby(purchase_data['SN']).count()
avg_purch_price = purchase_data['Price'].groupby(purchase_data['SN']).sum()/purch_count
total_purch_value = purchase_data['Price'].groupby(purchase_data['SN']).sum()

top_spender_df = pd.DataFrame({"Purchase Count": purch_count,
                              "Average Purchase Price" : avg_purch_price,
                              "Total Purchase Value" : total_purch_value})

top_spender_df.sort_values(by="Total Purchase Value",ascending=False,inplace=True)

top_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_items = purchase_data.groupby(["Item ID","Item Name"])

buy_count = grouped_items["SN"].count()
price_sum = grouped_items["Price"].sum()

item_df = pd.DataFrame({"Purchase Count":buy_count,
                       "Item Price": price_sum/buy_count,
                       "Total Purchase Value":price_sum})
item_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_df.sort_values(by="Total Purchase Value", ascending = False, inplace=True)
item_df.head()
```
